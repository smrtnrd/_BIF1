```python
#############################
# Retrieve NCBI Data Online #
#############################
genes = ['gyrA','gyrB','rpoB']
Entrez.email     = "bbitj026@uottawa.ca"             # Always tell NCBI who you are
for gene in genes:
    search_term = gene + '[gene] AND citrobacter braakii[orgn] AND complete genome[title]'
    handle = Entrez.esearch(db='nucleotide', term=search_term)
    genome_ids = Entrez.read(handle)['IdList']

###############################
# Generate Genome Fasta files #
###############################

    for genome_id in genome_ids:
        record = Entrez.efetch(db="nucleotide", id=genome_id, rettype="gb", retmode="text")

        filename = 'genBankRecord_{}_{}.gb'.format(genome_id, gene)
        print('Writing:{}'.format(filename))
        with open(filename, 'w') as f:
            f.write(record.read())
    

```
